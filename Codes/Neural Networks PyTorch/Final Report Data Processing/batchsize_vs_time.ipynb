```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
number_of_batches = [1, 25, 50, 75, 100]  # x-axis: number of batches
time_taken = [500,100,50,20,10]  # y-axis: time taken (random data for illustration)

# Create line plot
plt.figure(figsize=(10,4))
plt.plot(number_of_batches, time_taken, 'b.-') 
plt.xlabel('Batch Size', fontsize = 13)
plt.ylabel('Time Taken (seconds)', fontsize = 13)
plt.title('Time Taken vs Batch Size', fontsize = 13)
plt.grid(True, alpha=0.5)
plt.show()

```
