```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-7, 7, 1000)
y = sigmoid_activation(x)

plt.figure(figsize=(7, 5))

plt.axvline(x=0, color='grey') # add a vertical line at x=0
plt.axhline(y=0, color='grey', linestyle="--", linewidth=1) # add a vertical line at x=0
plt.axhline(y=1, color='grey', linestyle="--", linewidth=1) # add a vertical line at x=0
plt.axhline(y=0.5, color='grey', linestyle="--", linewidth=1) # add a vertical line at x=0
plt.tick_params(which='both', direction='in', top=True, right=True)
plt.plot(x, y, 'b',linewidth=2)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.savefig('sinh_activation.pdf', format='pdf')
plt.show()

```
